```python
%matplotlib inline

from ruamel.yaml import YAML
import numpy as np
import matplotlib.pyplot as plt

with open("./result.yaml") as f:
    data = YAML().load(f.read())

xs = np.array(data["xs"])
ys = np.array(data["ys"])
ws = np.array(data["ws"])
bs = np.array(data["bs"])
true_w = np.array(data["true_w"])
true_b = np.array(data["true_b"])

def plot_line(w, b, lw, s, c=None):
    xs = np.arange(-40, 40, 1)
    ys = w * xs + b
    if c is not None:
        plt.plot(xs, ys, linewidth=lw, linestyle=s, color=c)
    else:
        plt.plot(xs, ys, linewidth=lw, linestyle=s)

plt.figure(figsize=(10, 10))
plt.scatter(xs, ys)
plot_line(true_w, true_b, 4, "-", c="r")

for w, b in zip(ws, bs):
    plot_line(w, b, 0.25, "--")
```
